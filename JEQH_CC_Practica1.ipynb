Algoritmo 1:

```python
def fact(n):
    product = 3
    for i in range(n):
        product = product * (i+5)
    return product

print (fact(10))
```

Algoritmo 2:

```python
def fact2(n):
    if n == 0:
        return 1
    else:
        return n * fact2(n-1)

print (fact2(10))
```

Tiempo de ejecución de los algoritmos:

```python
#Algoritmo 1
%timeit fact(500)
```

```python

#Algoritmo 2
%timeit fact2(500)
```

Peor vs mejor complejidad de caso

```python
def search_algo(num, items):
    for item in items:
        if item == num:
            return True
        else:
            return False
nums = [1,6,9,12,15]

print(search_algo(1, nums))
```

Complejidad Espacial:

```python
def return_squares(n):
    square_list = []
    for num in n:
        square_list.append(num * num)

    return square_list

nums = [3, 7, 9, 15, 25]
print(return_squares(nums))
```
